```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
file_path = "./Resources/energy_data.csv"
energy_df = pd.read_csv(file_path)
energy_df.head()
```

```python
energy10_19_df = energy_df.loc[energy_df["Year"] >= 2010]
energy10_19_df.head()
```

```python
energy10_19_df.isnull().any()
```

```python
energy10_19_df.drop("Production.Coal", axis=1, inplace=True)
energy10_19_df.head()
```

```python
# Create a df for commercial, electric power, industrial, and residential consumption
CEIR_energy_df = pd.DataFrame({"State": energy10_19_df["State"],
                               "Year": energy10_19_df["Year"],
                   "Commercial_Coal": energy10_19_df["Consumption.Commercial.Coal"],
                   "Commercial_Natural_Gas": energy10_19_df["Consumption.Commercial.Natural Gas"],
                   "Commercial_Petroleum": energy10_19_df["Consumption.Commercial.Petroleum"],
                   "Commercial_Distillate_Fuel_Oil": energy10_19_df["Consumption.Commercial.Distillate Fuel Oil"],
                   "Commercial_Kerosene": energy10_19_df["Consumption.Commercial.Kerosene"],
                   "Commercial_Wood": energy10_19_df["Consumption.Commercial.Wood"],
                   "Commercial_Solar": energy10_19_df["Consumption.Commercial.Solar"],
                   "Commercial_Wind": energy10_19_df["Consumption.Commercial.Wind"],
                   "Commercial_Geothermal": energy10_19_df["Consumption.Commercial.Geothermal"],
                   "Commercial_Hydropower": energy10_19_df["Consumption.Commercial.Hydropower"],
                       "Electric_Power_Coal": energy10_19_df["Consumption.Electric Power.Coal"],
                       "Electric_Power_Natural Gas": energy10_19_df["Consumption.Electric Power.Natural Gas"],
                       "Electric_Power_Distillate_Fuel_Oil": energy10_19_df["Consumption.Electric Power.Distillate Fuel Oil"],
                       "Electric_Power_Wood": energy10_19_df["Consumption.Electric Power.Wood"],
                           "Industrial_Coal": energy10_19_df["Consumption.Industrial.Coal"],
                           "Industrial_Natural_Gas": energy10_19_df["Consumption.Industrial.Natural Gas"],
                           "Industrial_Petroleum": energy10_19_df["Consumption.Industrial.Petroleum"],
                           "Industrial_Distillate_Fuel_Oil": energy10_19_df["Consumption.Industrial.Distillate Fuel Oil"],
                           "Industrial_Kerosene": energy10_19_df["Consumption.Industrial.Kerosene"],
                           "Industrial_Wood": energy10_19_df["Consumption.Industrial.Wood"],
                           "Industrial_Solar": energy10_19_df["Consumption.Industrial.Solar"],
                           "Industrial_Wind": energy10_19_df["Consumption.Industrial.Wind"],
                           "Industrial_Geothermal": energy10_19_df["Consumption.Industrial.Geothermal"],
                           "Industrial_Hydropower": energy10_19_df["Consumption.Industrial.Hydropower"],
                                    "Residential_Coal": energy10_19_df["Consumption.Residential.Coal"],
                                    "Residential_Natural_Gas": energy10_19_df["Consumption.Residential.Natural Gas"],
                                    "Residential_Petroleum": energy10_19_df["Consumption.Residential.Petroleum"],
                                    "Residential_Distillate_Fuel_Oil": energy10_19_df["Consumption.Residential.Distillate Fuel Oil"],
                                    "Residential_Kerosene": energy10_19_df["Consumption.Residential.Kerosene"],
                                    "Residential_Wood": energy10_19_df["Consumption.Residential.Wood"]})
CEIR_energy_df.head()
```

```python
CEIR_energy_df.to_csv('./Resources/CEIR_energy.csv', index=False)
```

```python
CEIR_energy_df.to_json('./Resources/CEIR_energy.json', indent = 1, orient='records')
```
